```python
import pandas as pd
import re
```

```python
file_path = '/content/drive/MyDrive/Icon Conference/Data/Val_Task_A.xlsx'
data_val_task_a = pd.read_excel(file_path)
data_val_task_a.head()
```

```python
def clean_text(text):
    if not isinstance(text, str):
        return ''
    text = str(text)
    text = text.lower()
    text = re.sub(r'<br>', ' ', text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'@\w+|#+', '', text)
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text)

    return text.strip()
```

```python
original_string = data_val_task_a['Tweet'][0]
print("Original String: "+original_string)

cleaned_string = clean_text(original_string)
print("Cleaned String: "+cleaned_string)
```

```python
# Apply cleaning
data_val_task_a['Tweet'] = data_val_task_a['Tweet'].apply(clean_text)

print(data_val_task_a.head())

# Save cleaned data to CSV
cleaned_file_path = '/content/drive/MyDrive/Icon Conference/Data/Cleaned_Val_Task_A.csv'
data_val_task_a.to_csv(cleaned_file_path, index=False)
```
