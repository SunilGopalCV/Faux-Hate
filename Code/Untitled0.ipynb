```python
import matplotlib.pyplot as plt

# Data for epochs, fake macro F1 scores, and hate macro F1 scores
epochs = [1, 2, 3, 4, 5, 6, 7]
fake_f1_scores = [0.75, 0.78, 0.79, 0.79, 0.80, 0.79, 0.78]
hate_f1_scores = [0.70, 0.73, 0.76, 0.73, 0.75, 0.77, 0.75]

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(epochs, fake_f1_scores, label='Fake Macro F1', marker='o', linestyle='-', color='b')
plt.plot(epochs, hate_f1_scores, label='Hate Macro F1', marker='o', linestyle='-', color='r')

# Adding titles and labels
plt.title('Epoch vs Macro F1 Scores for Fake and Hate (Task A)')
plt.xlabel('Epoch')
plt.ylabel('Macro F1 Score')

# Adding grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
