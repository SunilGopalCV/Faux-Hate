```python
import pandas as pd

train_a = pd.read_csv('/content/cleaned_train.csv')
val_a = pd.read_csv('/content/cleaned_val.csv')

(train_a.head(), val_a.head())
```

```python
# prompt: give me a statistics on how many hate, non-hate, fake and non-fake are there in the tweet data on train_a as well as val_a dataset. It can be found using Hate and Fake colulmn which contains 0 or 1.

# Hate statistics for train_a
train_a_hate_stats = train_a['Hate'].value_counts()
print("Train_a Hate Statistics:")
print(train_a_hate_stats)

# Fake statistics for train_a
train_a_fake_stats = train_a['Fake'].value_counts()
print("\nTrain_a Fake Statistics:")
print(train_a_fake_stats)


# Hate statistics for val_a
val_a_hate_stats = val_a['Hate'].value_counts()
print("\nVal_a Hate Statistics:")
print(val_a_hate_stats)

# Fake statistics for val_a
val_a_fake_stats = val_a['Fake'].value_counts()
print("\nVal_a Fake Statistics:")
val_a_fake_stats
```
